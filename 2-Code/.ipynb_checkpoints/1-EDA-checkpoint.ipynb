````python
#Function that explores the null values
def df_EDA(df):
    print('DATAFRAME SHAPE:', df.shape)
    print('----------------')
    print('SUM OF NULL VALUES:', df.isnull().sum())
    print('----------------')
    print('DATA TYPES:')
    print(df.dtypes)
    print('----------------')
    print('DESCRIPTIVE STATISTICS:')
    return df.describe().T
````

### Dropping and tunnign features

Variables that show to be good predictors in both datasets are: `reanalysis_specific_humidity_g_per_kg`, `reanalysis_dew_point_temp_k`, `station_avg_temp_c`, `station_min_temp_c`

Variables that show to be less correlated in both datasets are: 
`reanalysis_tdtr_k`,`ndvi_n`, `reanalysis_max_air_temp_k`, `ndvi_se`, `station_diur_temp_rng_c`, `ndvi_nw`. 

Even if `year` and `week of the year` are not highly correlated, I will keep them giving that are the timeframe. 


````python
sj_train.drop('reanalysis_tdtr_k', axis=1, inplace=True)
iq_train.drop('reanalysis_tdtr_k', axis=1, inplace=True)

sj_train.drop('ndvi_ne', axis=1, inplace=True)
iq_train.drop('ndvi_ne', axis=1, inplace=True)

sj_train.drop('reanalysis_max_air_temp_k', axis=1, inplace=True)
iq_train.drop('reanalysis_max_air_temp_k', axis=1, inplace=True)

sj_train.drop('ndvi_se', axis=1, inplace=True)
iq_train.drop('ndvi_se', axis=1, inplace=True)

sj_train.drop('station_diur_temp_rng_c', axis=1, inplace=True)
iq_train.drop('station_diur_temp_rng_c', axis=1, inplace=True)

sj_train.drop('ndvi_nw', axis=1, inplace=True)
iq_train.drop('ndvi_nw', axis=1, inplace=True)
````
